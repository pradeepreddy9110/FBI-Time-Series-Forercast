## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Apply wrangling to both datasets
for df in [train, test]:
    # Drop duplicate rows (train only)
    if df is train:
        print("Original shape (train):", df.shape)
        df.drop_duplicates(inplace=True)
        print("After dropping duplicates (train):", df.shape)

    # Handle missing NEIGHBOURHOOD - check if column exists first
    if 'NEIGHBOURHOOD' in df.columns:  # Check if column exists
        df['NEIGHBOURHOOD'] = df['NEIGHBOURHOOD'].fillna('Unknown')
    
    # Drop rows with missing time values (HOUR, MINUTE) - check if columns exist
    if 'HOUR' in df.columns and 'MINUTE' in df.columns:
        df.dropna(subset=['HOUR', 'MINUTE'], inplace=True)

    # Drop rows with invalid or missing date - check if column exists
    if 'Date' in df.columns:
        df = df[df['Date'].notna()]

        # Convert 'Date' to datetime format
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

        # Feature Engineering: Extract time-based features
        df['weekday'] = df['Date'].dt.weekday
        df['week'] = df['Date'].dt.isocalendar().week
        df['year_month'] = df['Date'].dt.to_period('M').astype(str)

    # Clean TYPE (train only)
    if 'TYPE' in df.columns:
        df['TYPE'] = df['TYPE'].str.strip()
````

### What all manipulations have you done and insights you found?
